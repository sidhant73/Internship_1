Pre-Processing

```python
train_data = heart.copy()
```

```python
train_data["ChestPainType"].unique()
```

```python
print(train_data["RestingECG"].unique())
train_data["ST_Slope"].unique()
```

One-Hot Encoding

```python
train_data = pd.get_dummies(train_data, columns = ["RestingECG"], dtype = int)
train_data = pd.get_dummies(train_data, columns = ["ChestPainType"], dtype = int)
train_data = pd.get_dummies(train_data, columns = ["ST_Slope"], dtype = int)
train_data = pd.get_dummies(train_data, columns = ["Sex"], dtype = int)
```

```python
sns.heatmap(train_data.corr(), cmap="Blues")
```

```python
train_data.head()
```

```python
train_data.pop('ST_Slope_Flat')
train_data.pop('RestingECG_Normal')
train_data.pop('Sex_M')
train_data.pop('ChestPainType_TA')
```

```python
train_data["ExerciseAngina"] = train_data["ExerciseAngina"].map({'N':0, 'Y':1})
```

```python
train_data.head(5)
```

```python
x = train_data
```

```python
y = x.pop('HeartDisease')
```

```python
y
```

Scikit-Learn Start

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

Splitting Data into train test

```python
x_test, x_train, y_test, y_train = train_test_split(x,y, train_size = 0.8, shuffle = True)
```

Making Model

```python
model = GaussianNB() #making model on GaussianNB()
```

```python
model.fit(x_train, y_train) #fitting our model
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
print(classification_report(y_test, y_pred))
```
